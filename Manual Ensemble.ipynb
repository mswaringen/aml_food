```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
path = 'results'
all_files = glob.glob(path + "/*.csv")

df = pd.read_csv(r'data/sample.csv',usecols=['img_name'])

for f in all_files:
    col_names=['img_name', f[8:-4]] 
    df_tmp = pd.read_csv(f,header=0, names=col_names)
    df = pd.merge(df, df_tmp, how='inner')
    
df
```

```python
mode = df.mode(axis=1)

df['majority'] = np.where(mode.isna().any(1), mode[0], df.iloc[:,1])
df
```

```python
df['1=2'] = np.where(df.iloc[:,1] == df.iloc[:,2], True, False)
df['1=3'] = np.where(df.iloc[:,1] == df.iloc[:,3], True, False)
df['2=3'] = np.where(df.iloc[:,1] == df.iloc[:,2], True, False)
df['unanimous'] = np.where((df.iloc[:,1] == df.iloc[:,2]) & (df.iloc[:,1] == df.iloc[:,2]), True, False)


df
```
